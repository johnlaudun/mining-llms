# Hand Inspecting Responses

```python
import json
import textwrap
```

```python
# Read file into one big text:
with open('r-di-11.json', 'r') as f:
    data = json.load(f)

texts = [entry['text'] for entry in data]
for text in texts:
    print(textwrap.fill(text, width=70))
    print()  # Print a newline between texts
```
